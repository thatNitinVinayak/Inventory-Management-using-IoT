```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
items=pd.read_csv('E:/final year project/items.csv')
```

```python
items=items[['id','name','brand','categories','primaryCategories','prices.amountMin']]
```

```python
items.head()
```

```python
items.columns=['id','name','brand','category','product','price']
items.head()
```

```python
items['price'] = items['price'].apply(lambda x:x.split())
```

```python
items.head()
```

```python
items.head()
```

```python
#Using The min-max feature scaling:
df_min_max_scaled = items.copy()
  
# apply normalization techniques by Column 1
column = 'price'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
display(df_min_max_scaled)
```

```python
 items=df_min_max_scaled.copy()
```

```python
items.head()
```

```python
items['tags']=items['brand']+items['category']+items['product']
```

```python
items.head()
```

```python
items=items.drop(columns=['brand','category','product','price'])
```

```python
items.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
items['tags'] = items['tags'].apply(lambda x: " ".join(x))
```

```python
vector = cv.fit_transform(items['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
items[items['name']=='UBD-M9500 HDR UHD Upscaling Blu-ray Disc Player'].index[0]
```

```python
def recommend(item):
    index = items[items['name'] == item].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(items.iloc[i[0]].name)
```

```python
recommend('UBD-M9500 HDR UHD Upscaling Blu-ray Disc Player')
```
